The purpose of this project is to assess whether more individuals ride the subway when it rains and to generate a linear model to predict the number of hourly entries given the features provided in the dataset.

To determine the appropriate test to use, we can plot the data as a histogram. Because the distribution is not normal, we choose to use a non-parametric test, specifically the Mann-Whitney U.

```python
no_rain = subway.ENTRIESn_hourly[subway.rain == 0]
rain = subway.ENTRIESn_hourly[subway.rain == 1]
```

Welch's T-Test 2-sided
Positive t-value indicates rain is greater than no rain. If we are performing a one sided test, the p value would be half. In our case, it would be reasonable to use a one sided test, because in our original question, we are specifically interested in whether subway ridership increases. If we were simply interested in whether there was a difference, we would use a 2-sided test.

```python
scipy.stats.ttest_ind(rain, no_rain, equal_var=False)
```

#Mann Whitney U Test
This test is more robust and appropriate given the distribution of the data.

```python
scipy.stats.mannwhitneyu(rain, no_rain)
```

To address our first question, we perform both Welch's T-test and the Mann-Whitney U Rank test. Given that the data is not normally distributed, the Mann-Whitney U would be more appropriate as it is more robust. However, both of the tests return a p value of less than .05, and therefore our result is statistically significant and the probability of our data/observation given there is no difference is less than 10^-5. Our analysis showed that for each data point, there is an average of 1845 people when there is no rain, and an average of 2028 when it rains. This is a an increase of 183 on average, or approximately 10%, when it rains.

Model #1: No Weather Features 

```python
features_base = subway[['weekday']] ##Weekday is a binary variable
values = subway[['ENTRIESn_hourly']]

dummy_hour = pd.get_dummies(subway['hour'], prefix = 'hour') #Hourly is changed to a categorical variable
features_base = features_base.join(dummy_hour)

dummy_units = pd.get_dummies(subway['UNIT'], prefix = 'unit')
features_base = features_base.join(dummy_units)

dummy_day = pd.get_dummies(subway['day_week'], prefix = 'day')
features_base = features_base.join(dummy_day)

features_array = features_base.values
values_array = values.values

features_array_train, features_array_test, values_array_train, values_array_test = sklearn.cross_validation.train_test_split(features_array, values_array, random_state = 1)

print "Size of Training Data:",
print features_array_train.size
print "Size of Test Data:",
print features_array_test.size

model_base = sklearn.linear_model.LinearRegression()
model_base.fit(features_array_train, values_array_train)

print "Training Data Accuracy:",
print (model_base.score(features_array_train, values_array_train))
print "Test Data Accuracy:",
print (model_base.score(features_array_test, values_array_test))
```

Model #2: Only Weather Features

```python
features_weather = subway[['rain', 'precipi', 'fog', 'meantempi']]
values = subway[['ENTRIESn_hourly']]

features_array = features_weather.values
values_array = values.values

features_array_train, features_array_test, values_array_train, values_array_test = sklearn.cross_validation.train_test_split(features_array, values_array, random_state = 1)

model_base = sklearn.linear_model.LinearRegression()
model_base.fit(features_array_train, values_array_train)

print "Training Data Accuracy:",
print (model_base.score(features_array_train, values_array_train))
print "Test Data Accuracy:",
print (model_base.score(features_array_test, values_array_test))
```

Model#3: All Relevant Features

```python
features_weather = subway[['rain', 'precipi', 'fog', 'meantempi', 'weekday']]
values = subway[['ENTRIESn_hourly']]

dummy_hour = pd.get_dummies(subway['hour'], prefix = 'hour')
features_weather = features_weather.join(dummy_hour)

dummy_units = pd.get_dummies(subway['UNIT'], prefix = 'unit')
features_weather = features_weather.join(dummy_units)

dummy_day = pd.get_dummies(subway['day_week'], prefix = 'day')
features_weather = features_weather.join(dummy_day)

dummy_conds = pd.get_dummies(subway['conds'], prefix = 'condition')
features_weather = features_weather.join(dummy_conds)

features_array = features_weather.values
values_array = values.values

features_array_train, features_array_test, values_array_train, values_array_test = sklearn.cross_validation.train_test_split(features_array, values_array, random_state = 1)

model_base = sklearn.linear_model.LinearRegression()
model_base.fit(features_array_train, values_array_train)

print "Training Data Accuracy:",
print (model_base.score(features_array_train, values_array_train))
print "Test Data Accuracy:",
print (model_base.score(features_array_test, values_array_test))
```

#Final Model by Combining Station and Hour

```python
subway['placetime'] = subway.UNIT + ' ' + subway.hour.astype('str')
subway.columns
subway['placetime'].head()
```

```python
features_weather = subway[['rain', 'precipi', 'fog', 'meantempi', 'weekday']]


dummy_day = pd.get_dummies(subway['day_week'], prefix = 'day')
features_weather = features_weather.join(dummy_day)

dummy_pdt = pd.get_dummies(subway['placetime'], prefix = 'pdt')
features_weather = features_weather.join(dummy_pdt)

features_array = features_weather.values
values_array = subway.ENTRIESn_hourly.values

lm = sklearn.linear_model.LinearRegression()
lm.fit(features_array, values_array)
lm.score(features_array,values_array)
```

In our analysis, we provide evidence that on days where it rains, there is an slight increase in the number of individuals who ride the subway. This is based on Mann-Whitney U Test performed on the data. Despite being statistically significant, this information provides a limited amount of information about overall subway ridership. Our analysis showed that for each data point, there is an average of 1845 people when there is no rain, and an average of 2028 when it rains. This is a an increase of 183 on average, or approximately 10%, when it rains, but this difference is very small compared to the effects of station and time as depicted in our linear model.

Our assessment of the data shows that the weather does affect the number of individuals who travel on the subway. To better understand the affects of weather on subway ridership, we use information about the station, the time, and the weather data from the subway data to generate a model. Based on our analysis, the addition of addition weather features provides only marginal gains to our model. Thus we can conclude that although weather does play a role, the number of individuals that use the subway at a given place and time are largely independent of the weather.


#Reflection

In this instance, I believe that the data was sufficient to provide some insight into the NYC subway ridership. After the generating the preliminary models and visualizations, I reassess the factors that I believe contributed most to the model and attempted a final model that I believe would best predict our variable of interest. Ideally, I would have able to combine the place, time, and day features into a single feature that could be used to predict the the subway ridership for our data set. However, when attempting to do that, I received a memory error and instead chose to combine the two features that showed the most promise. The rationale for combining these 3 predictors together is that they are synergistic and therefore have interactions between the variables. The assumption for fitting a linear model is that the coefficient generated assumes all other factors are constant. However, in the case of subway data, one particular station may see a difference of 1000 people at a given station at a given day and time, while for another, there may be a difference of 10000. Therefore, it may be more beneficial to give a single prediction based on the combined factors of station, day, and time. The limitations of this however, are that given that we have only a month of data, for a particular instance of subway/day/time, we would only have approximately 4 data points. Therefore, in order to determine the effectiveness of the model, we would need more data and generate to generate a test set to prevent overfitting.

Nonetheless, with the remaining features, I generated a linear model with an R^2 value of .815. Due to the non-linearity of the data, this is significantly better than the previous model, where we look at station and time independently. In light of this, it may be plausible to simply use the conditional mean of entries/hour given station, date, and time to predict subway ridership given a set of predictors. Furthermore, the use of a linear model, also assumes that the variables of interest are normally distributed. In our case, we know that the variable we are attempting to predict is right-skewed, therefore it is possible that the predicted number of individuals riding the subway would be overestimated at the low end. This is most likely due to the fact that in our model, day is represented as an independent variable and the difference between a weekday and weekend is still very high. However, our model still explains a great deal of the variance associated with the data. To improve out model, it may be useful to transform the data with a log transformation.



Lastly, we look at the limitations of the data. In this particular instance, this data set represents observations of approximately 450 New York subway stations for a one month period. And although this data may generalize well, it is difficult to predict how well it will generalize outside this time period. We may see the say overall trend, however, the changing months/seasons, may also have some substantial effect on subway ridership.
